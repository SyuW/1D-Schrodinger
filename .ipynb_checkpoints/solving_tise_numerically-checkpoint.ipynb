```python
# Harmonic oscillator potential: p^2/2m + x^2/2
ho = lambda x: 0.5 * x ** 2

# infinite square well potential: 0 inside well, infinity outside
isw = lambda x: x * 0

# Triangular potential
def tp(k):
    return lambda x: abs(k * np.heaviside(x, 1) * x)

# Double well potential
dwp = lambda x: 5 * ((x - 1) ** 2) * ((x + 1) ** 2) * ((x - 2) ** 2) * ((x + 2) ** 2)

x_test = np.linspace(-10, 10, 500)
y_ho = ho(x_test)
y_isw = isw(x_test)
tri = tp(5)
y_lp = tri(x_test)
plt.plot(x_test, y_ho, color="red", label="Harmonic oscillator")
plt.plot(x_test, y_isw, color="blue", label="Infinite Square Well")
plt.plot(x_test, y_lp, color="green", label="Symmetric Linear")
plt.legend()
```

```python
# Trying to use the odeint package in Python

from scipy.integrate import odeint, simps


def get_system_state(state, x, E, V):
    """
    Get the state vector at value of x in order to solve the
    system of first-order ODES using scipy.integrate.odeint
    """
    
    y, z = state
    
    return np.array([z, 2 * (V(x) - E) * y])


def solve_se(xs, E, V, direction="F", dpsi=0.1):
    """
    Solve the 1D Time-independent Schrodinger Equation
        f''(x) = 2(V(x) - E)f(x)
        
    Where f(x) is the quantum wavefunction (AKA the 'state').
    To use scipy.integrate.odeint, turn this equation into a
    system of first order equations by defining
    
        y'(x) = z(x)
        z'(x) = g(x)y(x)
        
    where
        g(x) = 2(V(x) - E)
        
    For initial conditions (at xs[0]), we need y = 0 and z to
    be some small positive number, since the wavefunction decays
    to zero as x goes to plus/minus infinity
    """
    
    assert direction.upper() in ['F', 'B']
    
    # forward integration
    if direction.upper() == 'F':
        psi = odeint(get_system_state, [0.0, dpsi], xs, args=(E, V))[:,0]
    
    # backward integration
    if direction.upper() == 'B':
        psi = odeint(get_system_state, [0.0, -dpsi], xs[::-1], args=(E, V))[:,0][::-1]
        
    # normalization
    psi /= np.sqrt(simps(psi**2, x=xs))
    
    return psi


def check_matching_condition(x_vals, wfcn, E, pot_func, match_index):
    
    # matching index
    i_match = np.argmin(pot_func(x_vals))
    
    # find the classical turning points
    obj = pot_func(x_vals) - E
    idx = [ind[0] for ind in np.argwhere(np.diff(np.sign(obj))) if abs(obj[ind]) > 1e-6]
    turning_points = x_vals[idx]
    
    #scale = wfcn_right / wfcn_left
    #wfcn[:match_index] *= scale
    
    n = 0
    for i, val in enumerate(wfcn[1:]):
        if wfcn[i+1] * wfcn[i] < 0:
            print(f'Location of root: {x_vals[i]}')
            n += 1
    
    print(f'Node number: {n}')
```

```python
# find the energy eigenvalues using matrix methods and fixed map iterative process

n = 1000   # number of points
x = np.linspace(-10, 10, n)

# How many eigenstates do we want to find?
N_eigenstates_to_find = 30
eigenvalue_accuracy = 1e-7

# define second order differential operator
h = (x[2]-x[1])
off = np.ones(n-1)
on = np.ones(n)
Dsquared = 0.5 * (np.diag(off, k=1) - 2 * np.diag(on, k=0) + np.diag(off, k=-1)) / (h ** 2)

# single well potential
V = (abs(x) > 2) * 1000
# harmonic oscillator potential
V = ho(x)
# triangular well potential
# V = (x < 0) * 20 + (x >= 0) * 2 * x
# double well potential
# V = (abs(x) > 5) * 20 + (abs(x) < 2) * 10
# double level well potential
# V = (abs(x) > 5) * 40 + (abs(x) < 2) * (-10)
# inverted gaussian
# u = 10
# k = 0.1
# V = u * (-np.exp(-k * x**2) + 1)
# linear well
# k = 5
# V = k * abs(x)
# quartic oscillator
# V = 0.5 * x ** 4
# quartic double well
# V = 10 * (0.3 * x - 1) ** 2 * (0.3 * x + 1) ** 2
# anharmonic oscillator
# V = 0.25 * x ** 2 + 0.5 * x ** 4
# morse potential
# x = np.linspace(0.1, 10, int(1e3))
# a = 1
# x_e = 1
# V = 30 * (1 - np.exp(-a * (x - x_e))) ** 2
# linear ramp
# x = np.linspace(-10, 10, int(1e3))
# V = 2 * (x > 0) * x

# define the Hamiltonian matrix
H = -Dsquared + np.diag(V)
H_inv = np.linalg.inv(H)

# pre-allocate arrays for eigenvectors and eigenvalues
vs = np.zeros((N_eigenstates_to_find, n)) * np.nan
es = np.zeros(N_eigenstates_to_find) * np.nan

for jj in range(N_eigenstates_to_find):
    print(f'Solving for the {jj}-th eigenstate...')
    # Start with a random unit vector
    v = np.random.rand(n)
    v = v / np.linalg.norm(v)
    
    # initialize some variables
    E_previous = np.nan
    E = np.nan
    n_iterations = 0
    
    # Iterate until the eigenvalue converges to some value
    while not(abs(E - E_previous)/(E + E_previous) < eigenvalue_accuracy):
        # Remember the previous estimate of the eigenvalue
        E_previous = E
        # Apply the map v <-- A v
        v = H_inv.dot(v)
        # We can estimate the eigenvalue by looking at how the length
        # changed
        E = 1 / np.linalg.norm(v)
        v = E * v
        # Subtract the projections of this vector onto all previously found
        # eigenvectors
        for kk in range(jj):
            v -= vs[kk, :] * vs[kk, :].dot(v)
            
        # Just for fun, count how many times we iterate
        n_iterations += 1;

    # Make sure this is a unit vector before we store it
    v /= np.linalg.norm(v);
    # Store the newly found eigenvector and eigenvalue
    vs[jj, :] = v;
    es[jj] = E;
    # Announce our success:
    print(f'found in {n_iterations} iterations, with eigenvalue {E}')
```
